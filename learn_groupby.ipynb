```python
#7. Total Revenue and Number of Unique Customers per Region:
df.groupby('Region').agg({
    'TotalPrice':'sum',
    'CustomerID':'nunique'
})
```

```python
#8. Average Quantity per Order per Month (Ignoring Year):
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df.groupby(df['OrderDate'].dt.to_period('M'))['Quantity'].mean().reset_index()
#9. For each Region and Product Category, find the Total Quantity and Max Price Per Unit:
#10. Find the Customer who spent the most in each Region:
```

```python
#9. For each Region and Product Category, find the Total Quantity and Max Price Per Unit
df.groupby(['Region','ProductCategory']).agg({
    'Quantity':'sum',
    'PricePerUnit':'max'
})
```

```python
#10. Find the Customer who spent the most in each Region
res = df.groupby(['Region','CustomerID'])['TotalPrice'].sum()
res.groupby('Region').idxmax().reset_index()
```

```python
# **Advanced Level (10 Questions)** 
# 1. Percentage of Total Revenue Contributed by Each Product Category
res = df.groupby('ProductCategory')['TotalPrice'].sum().reset_index()
total_revenue = res['TotalPrice'].sum()
res['percentage'] = (res['TotalPrice']/total_revenue)*100
res
```

```python
# 2. For Each Order, Calculate the Average PricePerUnit for its ProductCategory
df['AvgCategoryPrice'] = df.groupby('ProductCategory')['PricePerUnit'].transform('mean')
df
```

```python
# 3. Identify Customers Whose Average Order Price is Above the Overall Average Order Price:
overall_avg = df['TotalPrice'].mean()
overall_avg
res = df.groupby('CustomerID')['TotalPrice'].mean().reset_index()
res[res['TotalPrice']>overall_avg]
```

```python
# 4. For Each Region and PaymentMethod, Find the Top 2 ProductCategory by Total Quantity Sold
grouped = df.groupby(['Region', 'PaymentMethod', 'ProductCategory'])['Quantity'].sum().reset_index()
top2 = (
    grouped
    .groupby(['Region', 'PaymentMethod'], group_keys=False)
    .apply(lambda x: x.nlargest(2, 'Quantity'))
    .reset_index(drop=True)
)
top2
```

```python
# 5. Calculate the Cumulative Sum of TotalPrice for each Customer over OrderDate:
ndf = df.sort_values(['CustomerID','OrderDate'])
ndf['CumulativeTotalPrice'] = ndf.groupby('CustomerID')['TotalPrice'].cumsum()
ndf
```

```python
# 6. Find the ProductName that appears most frequently in orders for each ProductCategory
res = df.groupby(['ProductCategory','ProductName'])['OrderID'].count().reset_index()
dd =res.sort_values(['OrderID'],ascending=False)
dd.groupby('ProductCategory').first().reset_index()
```

```python
order_span = df.groupby('CustomerID')['OrderDate'].agg(['min','max'])
order_span['new'] = (order_span['max'] - order_span['min']).dt.days
order_span
```

```python
# 8. Regions where at least one customer older than 60 years made a purchase:
res = df.groupby(['Region','CustomerAge']).value_counts().reset_index()
res[res['CustomerAge']>60].groupby('Region')['CustomerID'].max()
```

```python
# 8 Regions where at least one customer older than 60 years made a purchase:
age = df[df['CustomerAge']>60]
pd.Series(age['Region'].unique()).reset_index(name="Region")
```

```python
# 9. Create a new column 'OrderValueCategory' (Low, Medium, High) based on the customer's average order value within their ProductCategory
df['AvgOrderValue'] = df.groupby(['CustomerID','ProductCategory'])['TotalPrice'].transform('mean')
def classify_order(row):
    if row['TotalPrice'] < row['AvgOrderValue']:
        return 'Low'
    elif row['TotalPrice'] > row['AvgOrderValue']:
        return 'High'
    else:
        return 'Medium'

df['OrderValueCategory'] = df.apply(classify_order, axis=1)

df
```

```python
# 10. For each customer, find the product category they spent the most money on.
res = df.groupby(['CustomerID','ProductCategory'])['TotalPrice'].sum().reset_index()
sorted_grouped = res.sort_values(['CustomerID', 'TotalPrice'], ascending=[True, False])
top_category = sorted_grouped.groupby('CustomerID').first().reset_index()
top_category
```
